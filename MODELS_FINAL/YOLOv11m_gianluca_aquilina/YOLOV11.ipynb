```python
import numpy as np
from ultralytics import YOLO
import cv2
import os

model = YOLO('/content/drive/MyDrive/AdvnacedCV/yolov11_experiment/weights/best.pt')

test_images_path = "/content/drive/MyDrive/AdvnacedCV/dataset/test/images"
output_path = "/content/drive/MyDrive/AdvnacedCV/dataset/Output"
os.makedirs(output_path, exist_ok=True)

image_analytics = {}

#colours for classes
num_classes = len(model.names)
class_colors = {
    model.names[i]: tuple(
        int(c) for c in cv2.cvtColor(
            src=np.array([[[i * 180 // num_classes, 255, 255]]], dtype=np.uint8),
            code=cv2.COLOR_HSV2BGR
        )[0, 0]
    ) for i in model.names
}


for filename in os.listdir(test_images_path):
    if filename.endswith('.jpg'):
        image_path = os.path.join(test_images_path, filename)

        #inference
        results = model(image_path)

        image = cv2.imread(image_path)
        class_counts = {model.names[i]: 0 for i in model.names}

        #iter over detected obj
        for box in results[0].boxes:
            #bounding box
            x_min, y_min, x_max, y_max = map(int, box.xyxy[0])
            conf = box.conf[0].item()
            cls = int(box.cls[0].item())
            label = f"{model.names[cls]} {conf:.2f}"

            if model.names[cls] in class_counts:
                class_counts[model.names[cls]] += 1
            color = class_colors[model.names[cls]]

            #drawing bounding box
            cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, 2)
            cv2.putText(image, label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        image_analytics[filename] = class_counts

        #to show analytics on img
        y_offset = 30
        x_offset = image.shape[1] - 200

        for cls_name, count in class_counts.items():
            if count > 0:
                key_text = f"{cls_name}: {count}"
                cv2.putText(image, key_text, (x_offset, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 4)
                cv2.putText(image, key_text, (x_offset, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 140, 255), 2)

                y_offset += 25


        #save output img to folder
        output_file_path = os.path.join(output_path, filename)
        cv2.imwrite(output_file_path, image)

#print analytics
for img_name, counts in image_analytics.items():
    print(f"Image: {img_name}")
    for cls_name, count in counts.items():
        if count > 0:
            print(f"  {cls_name}: {count}")
```
