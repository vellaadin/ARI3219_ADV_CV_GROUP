```python
model = YOLO("./Models/yolov8s.pt")

# Train the model
model.train(data="./dataset/data.yaml", 
            epochs=50,
            imgsz=640,
            pretrained=False,
            )
```

```python
best_model = YOLO('./runs/detect/train/weights/best.pt')

metrics = best_model.val(
    data="./dataset/data.yaml",
    split="test",
    classes=[0, 1, 2, 3]
)

print(metrics)
```

```python
import random

img_path = "./dataset/test/images"
img_list = os.listdir(img_path)


# select 5 random images
random.shuffle(img_list)

img_list = img_list[:5]

class_colors = {
    0: {"color": (0, 255, 0), "label": "black bag"},
    1: {"color": (255, 0, 0), "label": "grey bag"},
    2: {"color": (0, 0, 255), "label": "other waste"},
    3: {"color": (255, 255, 0), "label": "white bag"}
}


for img in img_list:
    path = os.path.join(img_path, img)
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # detect objects in the image
    detections = best_model(path)

    no_bags = {0:0, 1:0, 2:0, 3:0}
    
    # show the image with the detected objects
    for detection in detections[0].boxes:
        detection = detection.cpu().numpy()
        x1, y1, x2, y2 = detection.xyxy[0]

        # convert float to int
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        confidence = detection.conf[0].item()
        cls = (int)(detection.cls[0].item())
        no_bags[cls] += 1
        class_color = class_colors[cls]["color"]
        image = cv2.rectangle(image, (x1, y1), (x2, y2), class_color)
        image = cv2.putText(image, f"{class_colors[(int)(detection.cls[0].item())]['label']} {confidence:.2f}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, class_color, 2)
    
    i = 0
    for n in no_bags:    
        if no_bags[n] > 0:
            img_height, img_width, _ = image.shape
            image = cv2.putText(image, f"{class_colors[n]['label']} bags: {no_bags[n]}", (10, img_height-30 - 30*i), cv2.FONT_HERSHEY_SIMPLEX, 0.9, class_colors[n]["color"], 2)
            i += 1

    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis('off')
    plt.title(f"Detections for {img}")
    plt.show()
```

```python
model = best_model
test_images_path = "./dataset/test/images"
output_path = r"./outputs"
os.makedirs(output_path, exist_ok=True)

image_analytics = {}

#colours for classes
num_classes = len(model.names)
class_colors = {
    model.names[i]: tuple(
        int(c) for c in cv2.cvtColor(
            src=np.array([[[i * 180 // num_classes, 255, 255]]], dtype=np.uint8),
            code=cv2.COLOR_HSV2BGR
        )[0, 0]
    ) for i in model.names
}


for filename in os.listdir(test_images_path):
    if filename.endswith('.jpg'):
        image_path = os.path.join(test_images_path, filename)

        #inference
        results = model(image_path)

        image = cv2.imread(image_path)
        class_counts = {model.names[i]: 0 for i in model.names}

        #iter over detected obj
        for box in results[0].boxes:
            #bounding box
            x_min, y_min, x_max, y_max = map(int, box.xyxy[0])
            conf = box.conf[0].item()
            cls = int(box.cls[0].item())
            label = f"{model.names[cls]} {conf:.2f}"

            if model.names[cls] in class_counts:
                class_counts[model.names[cls]] += 1
            color = class_colors[model.names[cls]]

            #drawing bounding box
            cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, 2)
            cv2.putText(image, label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        image_analytics[filename] = class_counts

        #to show analytics on img
        y_offset = 30 
        x_offset = image.shape[1] - 200

        for cls_name, count in class_counts.items():
            if count > 0:
                key_text = f"{cls_name}: {count}"
                cv2.putText(image, key_text, (x_offset, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 4)
                cv2.putText(image, key_text, (x_offset, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 140, 255), 2)
                
                y_offset += 25


        #save output img to folder
        output_file_path = os.path.join(output_path, filename)
        cv2.imwrite(output_file_path, image)

#print analytics
for img_name, counts in image_analytics.items():
    print(f"Image: {img_name}")
    for cls_name, count in counts.items():
        if count > 0:
            print(f"  {cls_name}: {count}")
```
